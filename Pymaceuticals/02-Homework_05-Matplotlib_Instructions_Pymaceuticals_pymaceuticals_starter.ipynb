```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
clinical_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
mouse_df=pd.DataFrame(mouse_data)
clinical_df=pd.DataFrame(clinical_data)

full_df=pd.merge(mouse_df,clinical_df,on="Mouse ID",how="outer")

# Display the data table for preview
full_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean = full_df.groupby(["Drug","Timepoint"]).mean()

# Convert to DataFrame
mean_tumor=pd.DataFrame(mean)
mean_tumor=mean_tumor.drop("Metastatic Sites",axis=1)
# Preview DataFrame
mean_tumor.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
sem = full_df.groupby(["Drug","Timepoint"]).sem()

# Convert to DataFrame
sem_tumor = pd.DataFrame(sem)
sem_tumor=sem_tumor.drop(["Metastatic Sites","Mouse ID"],axis=1)

# Preview DataFrame
sem_tumor.head()
```

```python
#Minor Data Munging to Re-Format the Data Frames
mean_tumor=mean_tumor.reset_index()
plot_data=mean_tumor.pivot(index='Timepoint',columns='Drug',values= 'Tumor Volume (mm3)')



#Preview that Reformatting worked
plot_data.head()
```

```python
# Generate the Plot (with Error Bars)
plt.figure(figsize = (10, 8))
plt.errorbar(x=plot_data.index,y=plot_data["Capomulin"],linestyle="--",fmt="o",markersize=5,linewidth=0.5,label="Capomulin",color="y")
plt.errorbar(x=plot_data.index,y=plot_data["Infubinol"],linestyle="--",fmt="^",markersize=5,linewidth=0.5,label="Infubinol",color="b")
plt.errorbar(x=plot_data.index,y=plot_data["Ketapril"],linestyle="--",fmt="s",markersize=5,linewidth=0.5,label="Ketapril",color="g")
plt.errorbar(x=plot_data.index,y=plot_data["Placebo"],linestyle="--",fmt="d",markersize=5,linewidth=0.5,label="Placebo",color="black")
plt.errorbar(x=plot_data.index,y=plot_data["Ceftamin"],linestyle="--",fmt=",",markersize=5,linewidth=0.5,label="Ceftamin",color="r")
plt.errorbar(x=plot_data.index,y=plot_data["Naftisol"],linestyle="--",fmt=".",markersize=5,linewidth=0.5,label="Naftisol",color="c")
plt.errorbar(x=plot_data.index,y=plot_data["Propriva"],linestyle="--",fmt="<",markersize=5,linewidth=0.5,label="Propriva",color="y")
plt.errorbar(x=plot_data.index,y=plot_data["Ramicane"],linestyle="--",fmt=">",markersize=5,linewidth=0.5,label="Ramicane",color="m")
plt.errorbar(x=plot_data.index,y=plot_data["Stelasyn"],linestyle="--",fmt="+",markersize=5,linewidth=0.5,label="Stelasyn",color="r")
plt.errorbar(x=plot_data.index,y=plot_data["Zoniferol"],linestyle="--",fmt="_",markersize=5,linewidth=0.5,label="Zoniferol",color="b")

x_lim=len(plot_data.index)
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.legend(loc="upper left")
plt.grid()
plt.show()

# Show the Figure
plt.show()

# Save the Figure
plt.savefig("Images/Tumor Respose to Treatment.png")
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
#mean_tumor=mean_tumor.reset_index()

plot_data=mean_tumor.pivot(index='Timepoint',columns='Drug',values= 'Tumor Volume (mm3)')
tumor_percent=((plot_data.iloc[-1]-plot_data.iloc[0])/plot_data.iloc[0])*100

tumor_percent
```

```python
# Store all Relevant Percent Changes into a Tuple
drug=list(tumor_percent.index)
percent=list(tumor_percent.values)

# Splice the data between passing and failing drugs
color=["r" if percent > 0 else "g" for percent in tumor_percent]
plt.figure(figsize = (10, 8))
plt.bar(drug,percent,color=color,align="edge")
plt.grid(linewidth =0.5)
plt.ylabel("% Tumor Volume Change")
plt.title("Tumor Change Over 45 days")
plt.savefig("Images/Summary.png")
```
